**Detecting and handling outliers**

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 36, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 120000, 90000, 90000]
}

df = pd.DataFrame(data)

# Detecting outliers using the IQR method
Q1 = df[['Age', 'Salary']].quantile(0.25)
Q3 = df[['Age', 'Salary']].quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
outliers = ((df[['Age', 'Salary']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Salary']] > (Q3 + 1.5 * IQR)))

# Display the DataFrame with outliers marked as True
print(outliers)
```

```python
# Remove Outliers
df_no_outliers = df[~((df[['Age', 'Salary']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Salary']] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_no_outliers)

# Replace Outliers
df_outliers_handled = df.copy()
for column in ['Age', 'Salary']:
    df_outliers_handled.loc[outliers[column], column] = df[column].median()  # or df[column].mean()
print(df_outliers_handled)
```

```python
roots = ["cat", "bat", "rat"]
sentence = "the cattle was rattled by the battery"


def replace_words(roots, sentence):
    words = sentence.split(" ")

    # looping over each word
    for index, word in enumerate(words):
        # looping over each root
        for root in roots:
            # checking if words start with root
            if word.startswith(root):
                # replacing the word with its root
                words[index] = root
    return " ".join(words)


replace_words(roots, sentence)


# 'the cat was rat by the bat'
```

```python
import re

def is_palindrome(text):
    # lowering the string
    text = text.lower()

    # Cleaning the string
    rx = re.compile('\W+')
    text = rx.sub('',text).strip()

    # Reversing and comparing the string
    return text == text[::-1]
```

```python
def is_palindrome(text):
    # lowering the string
    text = text.lower()

    # Cleaning the string
    rx = re.compile('\W+')
    text = rx.sub('',text).strip()

    # Reversing the string
    rev = ''.join(reversed(text))
    return text == rev
```
